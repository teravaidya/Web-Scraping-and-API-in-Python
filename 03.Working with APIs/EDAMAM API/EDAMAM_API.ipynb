Sending a POST request

```python
headers = {
    'Content-Type' : 'application/json'
}
```

```python
recipe = {
    'title' : 'Cappucinno',
    'ingr' : ['3 oz baileys irish cream', '5 oz hot coffee', 'dash(es) whipped cream (from a can)', 'dash(es) nutmeg']
}
```

```python
r = requests.post(url, headers = headers, json = recipe)
r.status_code
```

```python
capp_info = r.json()
print(json.dumps(capp_info, indent = 4))
```
